This is a LLM Call function which will be used for all the prompts 

```python
from ollama import chat

def llm_call(prompt: str) -> str:
    response = chat(model="llama3.2", messages=[{"role": "user", "content": prompt}])
    
    return response["message"]["content"]
```

Zero-Shot Prompting: In this technique, the model is given only the task instruction without any examples.

```python
zero_shot_prompt =f"""Generate a cover letter based on the following:
Resume:
{resume_uni}
Job: Looking for a backend developer proficient in Django at a tech startup."""
print("Zero-Shot Output:")
print(llm_call(zero_shot_prompt))
```

Chain Of Thought technique: In this technique model performs a task by guiding them through a step-by-step reasoning process. It involves generating intermediate reasoning steps to solve complex tasks.

```python
cot_promt = f"""
Generate a full cover letter based on the resume and role given. Include 
1.Header:name,github link,contact info and hiring manager role and company
2.Para 1:Address the hiring manager.
Express excitement about the role (mention job title and company).
Highlight 1 unique thing about the company that you connect with (shows research).
Provide a 1 or 2 -sentence summary of why you're a strong match (your standout value).
3.Para 2:Briefly describe a key project, internship, or job relevant to the role.
Link your experience to the specific requirements in the job post.
4.Para 3:Highlight a unique strength (e.g., leadership style, problem-solving mindset, adaptability).
Relate it to the company's mission, values, or work culture.
5.closing para:Reiterate excitement to contribute.
Mention availability for interviews.
Thank them for their time and express anticipation

Input:
Resume:{resume_uni}
Job: Looking for a backend developer proficient in Django at a tech startup.
Cover Letter:
"""
print("Chain of Thought Output:")
print(llm_call(cot_promt))
```

Meta Prompting : This techniqe involves creating instructions that help large language models generate more precise and focused prompts. It provides key information, examples, and context to build prompt components.

```python
meta_prmpt = f"""
Context: You an my ai assistant who are a highly skilled recruiter who writes my cover letter for my application
based on my resume and job role i am applying.
Instruction: you need to write a convincing and a recruitment worthy cover letter covering about the project I've worked on,skill set i have that provides value to the team
,also include skills that can be used in a team environment. Write it such a way that a recruiter is intrigued and calls for an interview(mention dates you are available to interview calls).
Resume:{resume_uni}
Job role: Looking for a backend developer proficient in Django at a tech startup.
Cover Letter:
"""
print("meta prompting output:")
print(llm_call(meta_prmpt))
```

Self Consistent Prompting:This technique involves generating multiple outputs for a single query and selecting the most consistent answer. It helps improve the accuracy of the model's responses by averaging multiple reasoning paths.

```python
self_consitent_prompt = f"""
You are an AI writing assistant. To ensure accuracy and consistency, internally generate at least three different versions of a personalized cover letter using the resume and job description below. Then, select the version that is the most consistent, relevant, and professional among them and present only that final version.

Resume:{resume_uni}

Job Description:
Looking for a backend developer proficient in Django and API development at a fast-paced tech startup focused on AI-powered products.

Now, generate the final consistent and refined cover letter.
"""
print("Self-Consistent prompting output:")
print(llm_call(self_consitent_prompt))
```

Prompt Chaining : This technique involves breaking down a complex task into a series of smaller, interconnected prompts. The output of one prompt serves as the input for the next, guiding the model through a structured reasoning process.

```python
#prompt_chaining technique
prompt_1 = f"Extract the key skills and experience from the following resume:\n{resume_uni}"
skills = llm_call(prompt_1)

job = "Backend developer role requiring Django and API development at a fast-growing tech startup."
prompt_2 = f"Match the following skills to this job description:\nSkills: {skills}\nJob: {job}"
matched_points = llm_call(prompt_2)

prompt_3 = f"Write a cover letter using these matched points:\n{matched_points}"
cover_letter = llm_call(prompt_3)
print("Prompt_chainning output:")
print(cover_letter)
```

General Knowledge Prompting: This method enhances the capability of models by providing contextual clues or prompts. It involves generating relevant prompts that guide AI systems toward producing more accurate and context-aware responses

```python
prompt_knowledge = """
Explain what startups typically expect from a backend developer in terms of skills, personality, and experience.
"""
knowledge = llm_call(prompt_knowledge)

prompt_final = f"""
Resume:{resume_uni}
Job: Backend developer at a startup.
Based on the following knowledge:\n{knowledge}\n
Generate a personalized cover letter.
"""
print("General_knowledge_prompting out")
print(llm_call(prompt_final))
```

Trees Of Thought: It works of the principle of Chain prompting, encourages exploration over thoughts that serve as intermediate steps for problem-solving. It uses a tree structure to explore multiple reasoning paths

```python
prompt_intro = f"""
Resume:
{resume_uni}

Write 3 different opening lines for a cover letter for a Python backend developer at a startup.
The intro should be professional, enthusiastic, and briefly highlight relevant experience.
"""
intros = llm_call(prompt_intro)


prompt_body = f"""
Resume:
{resume_uni}

List 3 different ways to describe backend experience using Django, REST APIs, and cloud infrastructure in a cover letter.
Focus on accomplishments, impact, and technologies.
"""
bodies = llm_call(prompt_body)


prompt_close = f"""
Resume:
{resume_uni}

Suggest 3 strong closing lines for a cover letter applying to a tech startup.
The tone should be confident, forward-looking, and show eagerness for the opportunity.
"""
closes = llm_call(prompt_close)


prompt_final = f"""
Based on the following options and resume, pick one from each section and compose a cohesive, professional cover letter.

Resume:
{resume_uni}
Intro:
{intros}
Body:
{bodies}
Conclusion:
{closes}
"""
print("Trees of thought output:")
final_letter = llm_call(prompt_final)
print(final_letter)
```

Active Prompting: Active prompting improves chain-of-thought prompting performance by selectively human-annotating exemplars where the model shows the most uncertainty. This approach focuses on the most challenging questions for the model.

```python
initial_prompt = f"""
Resume: {resume_uni}
Job: Python backend developer at a startup.
Write a cover letter for this role.
"""
draft = llm_call(initial_prompt)

check_prompt = f"""
Here is a cover letter:

{draft}

Please critique this letter. What are its strengths and weaknesses?
"""
feedback = llm_call(check_prompt)

improve_prompt = f"""
Based on the feedback below, improve the cover letter:
Feedback: {feedback}
Cover Letter: {draft}
give out the complete output
"""
final_version = llm_call(improve_prompt)
print("Active Prompting ouput:")
print(final_version)
```

ReAct prompting: ReAct combines reasoning and acting by prompting models to generate reasoning traces and task-specific actions in an interleaved manner. This allows the model to interact with external sources and update its knowledge dynamically.

```python
react_prompt = f"""
You are an AI assistant that writes cover letters by reasoning and acting step-by-step.
Goal: Write a tailored cover letter using the resume and job description.
Input:
Resume:
{resume_uni}
Job Description:
Backend developer role requiring Django and API development at a fast-growing tech startup.

Format:
Thought: [Your reasoning here]
Action: [Summarize or extract key details from resume/job description]
Observation: [Compare match / missing elements]
Thought: [Decide structure for the letter]
Final Answer: [Output the complete cover letter]
"""
print(llm_call(react_prompt))
```
